# pandas层次化索引

## 1. 创建多层行索引

### 1) 隐式构造

最常见的方法是给DataFrame构造函数的index参数传递两个或更多的数组

- Series也可以创建多层索引

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
s = Series(np.random.randint(0,150,size=6),
           index=pd.MultiIndex.from_product([['张三','李四','王五'],['期中','期末']]))
s
```

```python
s = Series(np.random.randint(0,150,size=6),
           index=pd.MultiIndex.from_arrays([['张三','张三','李四','李四','王五','王五'],['期中','期末','期中','期末','期中','期末']]))
s
```

除了行索引index，列索引columns也能用同样的方法创建多层索引

```python
# 多层， 多维， 一层表示一个维度
s['张三','期末']
```

```python
s['张三':'李四']
```

```python
s['张三':'赵六']
```

```python
s.loc['张三':'王五']
```

## 4. 索引的堆（stack）

- ``stack()``
- ``unstack()``

```python
df
```

```python
# 行索引转换成列索引
df.unstack()
```

```python
df.unstack(level=0)
```

```python
df2
```

```python
df2.stack(level=0)
```

【小技巧】使用stack()的时候，level等于哪一个，哪一个就消失，出现在行里。

【小技巧】使用unstack()的时候，level等于哪一个，哪一个就消失，出现在列里。

## 5. 聚合操作

【注意】

- 需要指定axis

- 【小技巧】和unstack()相反，聚合的时候，axis等于哪一个，哪一个就保留。

所谓的聚合操作：平均数，方差，最大值，最小值……

```python
s
```

```python
s.max()
```

```python
s.max(level=0)
```

```python
s.max(level=1)
```

```python
df2
```

```python
df2.mean(axis=1)
```

```python
df2.mean(axis=0).mean(level=0)
```

```python
df2.mean(axis=0).mean(level=1)
```

```python
df2.pow(2)
```

```python
df2.rank()
```
