## 2. 使用pd.merge()合并

merge与concat的区别在于，merge需要依据某一共同的行或列来进行合并

使用pd.merge()合并时，会自动根据两者相同column名称的那一列，作为key来进行合并。

注意每一列元素的顺序不要求一致

###  1) 一对一合并

````python
df1 = DataFrame({'id':[1,2,3],'python':[88,99,124],'math':[123,89,78]})
df1
````

````python
df2 = DataFrame({'id':[1,4,5],'java':np.random.randint(0,150,size=3),'en':np.random.randint(0,150,size=3)})
df2                
````

````python
pd.merge(df1,df2,on='id')
````

### 2) 多对一合并

````python
df3 = DataFrame({'id':[1,1,5],'java':np.random.randint(0,150,size=3),'en':np.random.randint(0,150,size=3)})
df3
````

````python
pd.merge(df1,df3,on='id')
````

### 3) 多对多合并

````python
df4 = DataFrame({'id':[1,2,1],'python':[88,99,124],'math':[123,89,78]})
display(df3,df4)
````

````python
pd.merge(df3,df4)
````

### 4) key的规范化

- 使用on=显式指定哪一列为key,当有多个key相同时使用

````python
df1
````

````python
df5 = DataFrame({'id':[1,2,4],'python':[88,98,124],'pysic':[123,89,78]})
display(df1,df5)
````

````python
pd.merge(df1,df5)
````

````python
pd.merge(df1,df5,on='id',suffixes=('_期中','_期末'))
````

````python
pd.merge(df1,df5,on='python',suffixes=('_A','_B'))
````

- 使用left_on和right_on指定左右两边的列作为key，当左右两边的key都不想等时使用

````python
df1
````

````python
df6 = DataFrame({'PYTHON':[88,99,123],'en':[114,116,79],'chme':[99,134,147]})
df6
````

````python
pd.merge(df1,df6,left_on='python',right_on='PYTHON')
````

## 3. 案例分析：美国各州人口数据分析

首先导入文件，并查看数据样本

````python
pop = pd.read_csv('./state-population.csv')
pop.shape
````

````python
pop.head()
````

````python
abbr = pd.read_csv('./state-abbrevs.csv')
abbr
````

````python
areas = pd.read_csv('./state-areas.csv')
areas
````

合并pop与abbrevs两个DataFrame，分别依据state/region列和abbreviation列来合并。

为了保留所有信息，使用外合并。

````python
pop2 = pd.merge(pop,abbr,how='outer',left_on='state/region',right_on='abbreviation')
pop2.shape
````

去除abbreviation的那一列（axis=1）

````python
pop2.head()
````

````python
# inplace 替换原来的对象
pop2.drop('abbreviation',axis=1,inplace=True)
````

````python
pop2.head()
````

查看存在缺失数据的列。

使用.isnull().any()，只有某一列存在一个缺失数据，就会显示True。

````python
pop2.isnull().any()
````

查看缺失数据

根据数据是否缺失情况显示数据，如果缺失为True，那么显示

````python
cond = pop2.isnull().any(axis=1)
pop2[cond]['state/region'].unique()
# Puerto Rico == PR
# USA = united state
````

````python
areas
````

找到有哪些state/region使得state的值为NaN，使用unique()查看非重复值

为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN！

记住这样清除缺失数据NaN的方法！

````python
(pop2['state/region'] == 'PR').sum()
````

````python
#  找到缩写等于PR 的洲的条件 [False,True,.....]
cond = pop2['state/region'] == 'PR'

pop2['state'][cond] = 'Puerto Rico'
````

````python
cond = pop2['state/region'] == 'USA'
cond
pop2['state'][cond] = 'United State'
````

````python
pop2.isnull().any()
````

````python
# 2544数据中只有20条为空
pop2.isnull().any(axis = 1).sum()
````

````python
pop2.dropna(inplace=True)
````

````python
pop2.isnull().any()
````

合并各州面积数据areas，使用左合并。

思考一下为什么使用外合并？



````python
pop2.head()
````

````python
areas
````

````python
pop3 = pd.merge(pop2,areas,on='state',how='outer')
pop3.shape
````

继续寻找存在缺失数据的列

````python
pop3.head()
````

````python
pop3.isnull().any()
````

我们会发现area(sq.mi)这一列有缺失数据，为了找出是哪一行，我们需要找出是哪个state没有数据

````python
cond = pop3.isnull().any(axis=1)
pop3[cond]
````

````python
a = areas['area (sq. mi)'].sum()
a
````

去除含有缺失数据的行

````python
pop3.fillna(value=a,inplace=True)
````

````python
pop3.isnull().any()
````

查看数据是否缺失

````python
pop3.to_csv('./usa_pop.csv')
````

找出2010年的全民人口数据,df.query(查询语句)

````python
pop3.head()
````

````python
pop4 = pop3.query("year == 2010 and ages =='total'")
pop4
````

对查询结果进行处理，以state列作为新的行索引:set_index

````python
pop4.set_index(keys = 'state',inplace=True)
````

计算人口密度。注意是Series/Series，其结果还是一个Series。

````python
density = pop4['population']/pop4['area (sq. mi)']
density
````

````python
# 将series——————》dataframe
df_density = DataFrame(density,columns=['pop_density'])
df_density
````

````python
display(pop4.head(),df_density.head())
````

````python
pop5 = pd.merge(pop4,df_density,left_index=True,right_index=True)
pop5
````

排序，并找出人口密度最高的五个州sort_values()

````python
pop5.sort_values(by = 'pop_density',ascending=False)
````

找出人口密度最低的五个州

要点总结：
- 统一用loc()索引
- 善于使用.isnull().any()找到存在NaN的列
- 善于使用.unique()确定该列中哪些key是我们需要的
- 一般使用外合并、左合并，目的只有一个：宁愿该列是NaN也不要丢弃其他列的信息

## 回顾：Series/DataFrame运算与ndarray运算的区别

- Series与DataFrame没有广播，如果对应index没有值，则记为NaN；或者使用add的fill_value来补缺失值
- ndarray有广播，通过重复已有值来计算
