```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv2
%matplotlib inline
```

```python
jin = cv2.imread('./jin.jpg')[:,:,::-1]
```

```python
jin.shape
```

```python
plt.imshow(jin)
```

```python
face_detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
face_zone = face_detector.detectMultiScale(jin)
face_zone
```

```python
dog = plt.imread('./d.jpg')
d_head = dog[40:450,240:730]
plt.imshow(d_head)
d_head.shape
```

```python
plt.figure(figsize=(1,1))
d_head = cv2.resize(d_head,(74,74))
plt.imshow(d_head)
```

```python
face_zone
```

```python
for x,y,w,h in face_zone:
    jin[y:y+h,x:x+w] = d_head
```
