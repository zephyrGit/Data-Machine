```python
def convert(x):
    if x < 60:
        return '不及格'
    elif (x >= 60) & (x < 80):
        return '及格'
    elif (x >= 80) & (x < 100):
        return '中等'
    elif (x >= 100) & (x < 120):
        return '良好'
    else:
        return '优秀'
```

## 5. 数据聚合【重点】

数据聚合是数据处理的最后一步，通常是要使每一个数组生成一个单一的数值。

数据分类处理：

 - 分组：先把数据分为几组
 - 用函数处理：为不同组的数据应用不同的函数以转换数据
 - 合并：把不同组得到的结果合并起来
 
数据分类处理的核心：
     groupby()函数

```python
df = DataFrame({'item':np.random.randint(0,3,size=100),
                'sailer':np.random.randint(0,3,size=100),
                'price':np.random.randint(2,8,size=100),
               'weight':np.random.randint(30,50,size=100)})
df['item'] = df['item'].map({0:'白菜',1:'萝卜',2:'青椒'})

df
```

```python
df2 = df.copy()
```

```python
def convert(item):
    if item == 4 :
        return '张大妈'
#     transform 和 map的功能是相同的
# 不同点，map中可以放方法，也可以是字典，transform中只能放方法
```

```python
df2['price'] = df2['price'].transform(convert)
df2
```

```python
df['sailer'] = df['sailer'].map({0:'张大妈',1:'李大妈',2:'赵大叔'})
```

```python
df
```

```python
df.groupby(by = ['item'])['price','weight'].mean()
```

```python
df.groupby(by = ['sailer','item'])['price'].min()
```

```python
df.groupby(by = ['sailer','item'])[['price']].max()
```

```python
df.groupby(by = ['sailer','item'])[['price']].std()
```

## 6.0 高级数据聚合

可以使用pd.merge()函数将聚合操作的计算结果添加到df的每一行  
使用groupby分组后调用加和等函数进行运算，让后最后可以调用add_prefix()，来修改列名

```python
df.groupby(['item'])[['price']].mean()
```

```python
# display.float_format 保留两位小数
pd.set_option('display.float_format',lambda x:'%0.2f'%(x))
```

```python
df
```

```python
price_mean = df.groupby(['item'])[['price']].mean()
price_mean.rename(mapper={'price':'price_mean'},axis=1,inplace=True)
price_mean
```

```python
df2 = pd.merge(df,price_mean,left_on='item',right_index=True).loc[(np.arange(100))]
```

### 可以使用transform和apply实现相同功能

在transform或者apply中传入函数即可

```python
# 100条数据
df.groupby(['sailer','item'])['price'].apply(np.mean)
```

```python
price_mean = df.groupby(['sailer','item'])[['price']].transform(np.mean).add_suffix('mean_')
price_mean
```

```python
pd.merge(df2.drop('price_mean',axis=1),price_mean,left_index=True,right_index=True)
```
